## Функционал поиска по табличным данным

```python
from pydantic import BaseModel, Field
from typing import Optional, Dict, Any
from typing import Union, List

class QueryCompanyData(BaseModel):
    """Позволяет искать информацию в корпоративных таблицах ОБИТ:
    офисы, контакты, сотрудники, корпоративы и активности.
    """

    table: str = Field(description="Название таблицы: 'Адреса офисов', 'HR работа с персоналом', 'База сотрудники', 'Корпоративы', 'Активности'")
    filters: Optional[Dict[str, Any]] = Field(default=None, description="Фильтры по колонкам, например {'Город':'Москва'} или {'Офис':'Лиговский'}")
    what_to_return: Optional[Union[str, List[str]]] = Field(default=None, description="Какие столбцы вернуть: строка или список строк, например 'Телефон' или ['ФИО','Контакт']")

    def process(self, session_id: str = None):
        try:
            df = load_table(self.table)
        except Exception as e:
            return f"Ошибка при загрузке таблицы: {e}"

        # применяем фильтры
        if self.filters:
            for k, v in self.filters.items():
                if k not in df.columns:
                    return f"В таблице '{self.table}' нет столбца '{k}'. Доступные: {', '.join(df.columns)}"
                df = df[df[k].astype(str).str.contains(str(v), case=False, na=False)]

        if df.empty:
            return f"По запросу ничего не найдено в таблице '{self.table}'."

        # выбираем нужные поля
        if self.what_to_return:
            if isinstance(self.what_to_return, str):
                cols = [c for c in df.columns if self.what_to_return.lower() in c.lower()]
            elif isinstance(self.what_to_return, list):
                cols = [c for c in df.columns if any(w.lower() in c.lower() for w in self.what_to_return)]
            else:
                cols = []
            if cols:
                df = df[cols]
        return df.to_dict(orient="records")
```

## Функционал создания и обновления профиля

```python
import json
import os
from pydantic import BaseModel, Field
from typing import Union

PROFILES_DIR = "user_profiles"
os.makedirs(PROFILES_DIR, exist_ok=True)


class CreateUserProfile(BaseModel):
    """Создает профиль нового пользователя. Можно пропустить заполнение, указав skip=True."""

    nickname: str | None = Field(default=None, description="Имя пользователя или ник")
    position: str | None = Field(default=None, description="Должность сотрудника")
    location: str | None = Field(default=None, description="Город или офис")
    hobbies: list[str] | str | None = Field(default=None, description="Интересы или виды спорта")
    status: str | None = Field(default=None, description="Сотрудник или кандидат")
    skip: bool = Field(default=False, description="Пропустить создание профиля")

    def process(self, session_id: str):
        profile_path = os.path.join(PROFILES_DIR, f"{session_id}.json")

        # если профиль уже есть и помечен как skipped — ничего не делаем
        if os.path.exists(profile_path):
            with open(profile_path, encoding="utf-8") as f:
                data = json.load(f)
            if data.get("skipped"):
                return "Ты уже решил не заполнять профиль — всё в порядке 🙂"

        # если пользователь явно пропускает или ничего не указал
        if self.skip or not any([self.nickname, self.position, self.location, self.hobbies, self.status]):
            profile = {
                "nickname": None,
                "position": None,
                "location": None,
                "hobbies": [],
                "status": None,
                "created": __import__('datetime').datetime.now().isoformat(),
                "updated": __import__('datetime').datetime.now().isoformat(),
                "skipped": True
            }
            with open(profile_path, "w", encoding="utf-8") as f:
                json.dump(profile, f, ensure_ascii=False, indent=2)
            return "Окей профиль можно заполнить позже — я помогу и без него!"

        # обычное создание профиля
        profile = self.model_dump()
        if isinstance(profile.get("hobbies"), str):
            profile["hobbies"] = [profile["hobbies"]]
        profile["created"] = profile["updated"] = __import__('datetime').datetime.now().isoformat()

        with open(profile_path, "w", encoding="utf-8") as f:
            json.dump(profile, f, ensure_ascii=False, indent=2)

        name = profile["nickname"] or "друг"
        return f"Профиль создан! Приятно познакомиться, {name}"



class UpdateUserProfile(BaseModel):
    """Обновляет профиль пользователя — например, при добавлении интересов или подписок."""

    nickname: str = Field(description="Имя пользователя")
    subscriptions: list[str] | None = Field(default=None, description="Темы, на которые пользователь подписывается")
    hobbies: list[str] | None = Field(default=None, description="Обновлённый список интересов")

    def process(self, session_id: str):
        profile_path = os.path.join(PROFILES_DIR, f"{session_id}.json")
        if not os.path.exists(profile_path):
            return "Профиль не найден. Создай новый через знакомство."
        profile = json.load(open(profile_path, "r", encoding="utf-8"))
        for key, value in self.model_dump(exclude_none=True).items():
            if isinstance(value, list):
                profile[key] = list(set(profile.get(key, []) + value))
            else:
                profile[key] = value
        profile["updated"] = __import__('datetime').datetime.now().isoformat()
        json.dump(profile, open(profile_path, "w", encoding="utf-8"), ensure_ascii=False, indent=2)
        return f"Профиль {profile['nickname']} обновлён."


class ReadUserProfile(BaseModel):
    """Возвращает данные профиля для подмешивания в контекст агента."""

    nickname: str = Field(description="Имя пользователя")

    def process(self, session_id: str):
        path = os.path.join(PROFILES_DIR, f"{session_id}.json")
        if not os.path.exists(path):
            return "Профиль не найден."
        return json.load(open(path, encoding="utf-8"))
```

```python
import json
from openai import OpenAI
from pydantic import BaseModel

client = OpenAI(
    base_url="https://rest-assistant.api.cloud.yandex.net/v1",
    api_key=os.getenv("YANDEX_API_KEY"),
    project=os.getenv("FOLDER_ID")
)

model = f"gpt://{os.getenv('FOLDER_ID')}/gpt-oss-120b/latest"


class Agent:
    def __init__(self, instruction, tools=[], session_id="default", model=model, tool_choice="auto"):
        self.instruction = instruction
        self.model = model
        self.tool_choice = tool_choice
        self.tool_map = {x.__name__: x for x in tools if issubclass(x, BaseModel)}
        self.tools = [
            self._create_tool_annot(x) for x in tools
        ]
        self.user_sessions = {}

    def _create_tool_annot(self, x):
        if issubclass(x, BaseModel):
            return {
                "type": "function",
                "name": x.__name__,
                "description": x.__doc__,
                "parameters": x.model_json_schema(),
            }

    def __call__(self, message, session_id="default"):
        s = self.user_sessions.get(session_id, {"last_reply_id": None, "history": []})
        s["history"].append({"role": "user", "content": message})
        
        # Подмешиваем RAG контекст
        context = fetch_context(message)
        if not context.strip():
            context = "Контекст по запросу не найден."
            
        prompt = (
            f"{self.instruction}\n\n"
            f"Контекст из базы (используй только эти данные, не выдумывай факты):\n"
            f"{context}\n\n"
            f"Запрос пользователя: {message}\n"
            f"Если информации нет — скажи об этом честно."
        )

        # Отправляем в модель
        res = client.responses.create(
            model=self.model,
            store=True,
            tools=self.tools,
            tool_choice=self.tool_choice,
            previous_response_id=s["last_reply_id"],
            input=prompt,
        )

        tool_calls = [item for item in res.output if item.type == "function_call"]
        if tool_calls:
            for call in tool_calls:
                fn = self.tool_map[call.name]
                obj = fn.model_validate(json.loads(call.arguments))
                result = obj.process(session_id)
                if not isinstance(result, str):
                    result = json.dumps(result, ensure_ascii=False)
                out = [{
                    "type": "function_call_output",
                    "call_id": call.call_id,
                    "output": result
                }]
                res = client.responses.create(
                    model=self.model,
                    input=out,
                    tools=self.tools,
                    previous_response_id=res.id,
                    store=True,
                )

        s["last_reply_id"] = res.id
        s["history"].append({"role": "assistant", "content": res.output_text})
        self.user_sessions[session_id] = s
        return res.output_text
```

```python
ALLOWED = {"b","strong","i","em","u","s","del","a","code","blockquote"}

def sanitize_html_for_telegram(text: str) -> str:
    if not text:
        return ""

    # <br> перенос строки
    text = re.sub(r"<\s*br\s*/?\s*>", "\n", text, flags=re.I)

    # убираем блочные теги
    text = re.sub(r"</?\s*(ul|ol|li|p|div|span|html|body|head|h[1-6]|table|tr|td|th|hr)\b[^>]*>", "", text, flags=re.I)
    text = re.sub(r"<!DOCTYPE[^>]*>", "", text, flags=re.I)

    # удаляем неподдерживаемые теги, но оставляем содержимое
    def strip_unknown_tags(m):
        tag = m.group(1).lower()
        if tag in ALLOWED:
            return m.group(0)
        return ""

    text = re.sub(r"</?([a-zA-Z0-9]+)(\s[^>]*)?>", strip_unknown_tags, text)

    # экранируем случайные одинокие угловые скобки
    text = text.replace("< ", "&lt; ").replace(" >", "&gt;").replace("<<", "&lt;&lt;").replace(">>", "&gt;&gt;")

    return text.strip()
```

```python
# функции для уведомлений

# формирование словаря {пользователь : активности, на которые он подписан}
def get_users_events_dict() -> dict:
    users = os.listdir(PROFILES_DIR)
    users_events_dict = {}

    for user in users:
        profile_path = os.path.join(PROFILES_DIR, user)
        with open (profile_path, "r", encoding="utf-8") as f:
            data = json.load(f)
            subscriptions = data['subscriptions']
            user_id, _ = os.path.splitext(user)
            users_events_dict[user_id] = subscriptions

    return users_events_dict

# рассылка напоминаний об активностях
async def send_event_reminders(bot: Bot) -> None:
    # для теста расскоментировать нижнюю строку и закомментировать следующую
    # now = datetime(2025, 11, 4, 17, 59)
    now = datetime.now()
    current_weekday = now.weekday()
    events_weekday = current_weekday + 1
    if events_weekday == 7:
        events_weekday = 0

    events_tomorrow = []
    events = pd.read_excel(EVENTS_PATH, engine='openpyxl')

    for _, row in events.iterrows():
        day_name = str(row.get("День недели", "")).strip()
        day = WEEKDAYS.get(day_name, -1)
        if day == events_weekday:
            events_tomorrow.append(row.to_dict())

    # если нет активностей, то выходим
    if not events_tomorrow:
        return

    # формируем и рассылаем уведы, если есть
    for event in events_tomorrow:
        try:
            event_activity = event['Активность '].replace("\n", " ").strip()
            event_time = event['Время'].replace("\n", " ").strip()
            event_place = event['Место'].replace("\n", " ").strip()
            event_address = event['Адрес'].replace("\n", " ").strip()
            event_host = event['Ответственный'].replace("\n", " ").strip()

            msg = (
                f"⚡ Напоминание!\n\nЗавтра {event_activity} в {event_time}.\n"
                f"📍 {event_place}\n"
                f"🏠 {event_address}\n"
                f"👤 Ответственный: {event_host}"
            )

            users_events = get_users_events_dict()
            for user, events in users_events.items():
                if event_activity in events:
                    await bot.send_message(int(user), msg)

        except Exception as e:
            print(f"Ошибка при обработке события: {e}")
```
