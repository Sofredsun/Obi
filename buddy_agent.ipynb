```python
# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ —Ç–µ–∫—Å—Ç–æ–≤ –∏—Ö —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö —Ñ–∞–π–ª–æ–≤
from glob import glob
import pandas as pd
import tiktoken

def get_tokenizer(model):
    try:
        return tiktoken.encoding_for_model(model)
    except:
        return tiktoken.encoding_for_model("gpt-5-nano") 

def get_token_count(filename):
    tokenizer = get_tokenizer(model)
    with open(filename, "r", encoding="utf8") as f:
        return len(tokenizer.encode(f.read()))

def get_file_len(filename):
    with open(filename, encoding="utf-8") as f:
        l = len(f.read())
    return l

d = [
    {
        "File": fn,
        "Tokens": get_token_count(fn),
        "Chars": get_file_len(fn),
    }
    for fn in list(glob("data/–¢–µ–∫—Å—Ç—ã/**/*.md", recursive=True))
]

df = pd.DataFrame(d)
df
```

```python
from openai import OpenAI
import os

client = OpenAI(
    base_url="https://rest-assistant.api.cloud.yandex.net/v1",
    api_key=os.getenv("YANDEX_API_KEY"),
    project=os.getenv("FOLDER_ID"),
)

VECTOR_STORE_ID = "fvt5u67lqjkgavj67idt"  # —Ç–≤–æ–π —É–∂–µ —Å–æ–∑–¥–∞–Ω–Ω—ã–π –∏–Ω–¥–µ–∫—Å

def fetch_context(query: str, k: int = 5, rewrite: bool = True, max_chars_per_snippet: int = 800) -> str:
    """–ë–µ—Ä—ë–º top-k —Ñ—Ä–∞–≥–º–µ–Ω—Ç–æ–≤ –∏–∑ vector_store –∏ –≥–æ—Ç–æ–≤–∏–º –∫–æ–º–ø–∞–∫—Ç–Ω—ã–π –∫–æ–Ω—Ç–µ–∫—Å—Ç."""
    srch = client.vector_stores.search(
        vector_store_id=VECTOR_STORE_ID,
        query=query,
        rewrite_query=rewrite,
    )
    parts = []
    for item in srch.data[:k]:
        # —É —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –æ–±—ã—á–Ω–æ –µ—Å—Ç—å content[0].text
        txt = ""
        if getattr(item, "content", None) and len(item.content) and getattr(item.content[0], "text", None):
            txt = item.content[0].text
        if not txt:
            continue
        if len(txt) > max_chars_per_snippet:
            txt = txt[:max_chars_per_snippet] + "‚Ä¶"
        src = getattr(item, "filename", "document")
        parts.append(txt)
    return "\n\n---\n\n".join(parts)

def rag_answer(query: str, system_instruction: str, model: str | None = None) -> str:
    """–î–µ–ª–∞–µ–º —á–µ—Å—Ç–Ω—ã–π RAG: –ø–æ–∏—Å–∫ ‚Üí –∫–æ–Ω—Ç–µ–∫—Å—Ç ‚Üí –æ–¥–∏–Ω –≤—ã–∑–æ–≤ –º–æ–¥–µ–ª–∏."""
    model = model or f"gpt://{os.getenv('FOLDER_ID')}/gpt-oss-120b/latest"
    context = fetch_context(query)
    if not context:
        context = "–ö–æ–Ω—Ç–µ–∫—Å—Ç –ø–æ –∑–∞–ø—Ä–æ—Å—É –Ω–µ –Ω–∞–π–¥–µ–Ω."

    prompt = (
        f"{system_instruction}\n\n"
        f"–ö–æ–Ω—Ç–µ–∫—Å—Ç –∏–∑ –≤–Ω—É—Ç—Ä–µ–Ω–Ω–µ–π –±–∞–∑—ã (–∏—Å–ø–æ–ª—å–∑—É–π —Ç–æ–ª—å–∫–æ —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ, –Ω–µ –≤—ã–¥—É–º—ã–≤–∞–π —Ñ–∞–∫—Ç—ã):\n"
        f"{context}\n\n"
        f"–í–æ–ø—Ä–æ—Å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è: {query}\n\n"
        f"–°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–π —è—Å–Ω—ã–π –æ—Ç–≤–µ—Ç. –ï—Å–ª–∏ –≤ –∫–æ–Ω—Ç–µ–∫—Å—Ç–µ –Ω–µ—Ç –æ—Ç–≤–µ—Ç–∞ ‚Äî —Ç–∞–∫ –∏ —Å–∫–∞–∂–∏."
        
    )

    res = client.responses.create(
        model=model,
        input=prompt,
        store=False
    )
    return res.output_text or ""

instruction = """
–¢—ã ‚Äî –∫–æ—Ä–ø–æ—Ä–∞—Ç–∏–≤–Ω—ã–π –±–∞–¥–¥–∏ –∫–æ–º–ø–∞–Ω–∏–∏ –û–ë–ò–¢  
–¢–≤–æ—è —Ü–µ–ª—å ‚Äî –ø–æ–º–æ–≥–∞—Ç—å —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∞–º —Ä–∞–∑–±–∏—Ä–∞—Ç—å—Å—è –≤ –ø—Ä–∞–≤–∏–ª–∞—Ö, –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ö –∏ –∂–∏–∑–Ω–∏ –∫–æ–º–ø–∞–Ω–∏–∏, 
–Ω–æ –¥–µ–ª–∞—Ç—å —ç—Ç–æ —Ç–µ–ø–ª–æ, –ø–æ–Ω—è—Ç–Ω–æ –∏ —Å –ª—ë–≥–∫–∏–º —á—É–≤—Å—Ç–≤–æ–º —é–º–æ—Ä–∞.

–û—Ç–≤–µ—á–∞–π –Ω–∞ –æ—Å–Ω–æ–≤–µ –∫–æ–Ω—Ç–µ–∫—Å—Ç–∞, –∫–æ—Ç–æ—Ä—ã–π —Ç–µ–±–µ –¥–∞—é—Ç (–≤–Ω—É—Ç—Ä–µ–Ω–Ω–∏–µ –¥–æ–∫—É–º–µ–Ω—Ç—ã).  
–ï—Å–ª–∏ –≤ –∫–æ–Ω—Ç–µ–∫—Å—Ç–µ –Ω–µ —Ö–≤–∞—Ç–∞–µ—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ ‚Äî –ø—Ä–µ–¥–ª–æ–∂–∏ —É—Ç–æ—á–Ω–∏—Ç—å –∏–ª–∏ –ø–æ–º–æ–≥–∏, –∫–∞–∫ –º–æ–∂–µ—à—å.

–ü–∏—à–∏ –¥—Ä—É–∂–µ–ª—é–±–Ω–æ, –≤ –∫–æ–Ω—Ü–µ –º–æ–∂–µ—à—å –∑–∞–¥–∞–≤–∞—Ç—å –æ–±—â–∏–π —É—Ç–æ—á–Ω—è—é—â–∏–π –≤–æ–ø—Ä–æ—Å,
"–†–∞—Å—Å–∫–∞–∑–∞—Ç—å —á—Ç–æ-—Ç–æ –µ—â–µ?".

–ù–∏–∫–æ–≥–¥–∞ –Ω–µ –≤—ã–¥—É–º—ã–≤–∞–π —Ñ–∞–∫—Ç—ã, –∫–æ—Ç–æ—Ä—ã—Ö –Ω–µ—Ç –≤ –∫–æ–Ω—Ç–µ–∫—Å—Ç–µ.  
–ï—Å–ª–∏ –∫–æ–Ω—Ç–µ–∫—Å—Ç –æ–≥—Ä–∞–Ω–∏—á–µ–Ω ‚Äî —á–µ—Å—Ç–Ω–æ —Å–∫–∞–∂–∏, —á—Ç–æ –¥–∞–Ω–Ω—ã—Ö –Ω–µ—Ç, –Ω–æ –ø–æ—Å—Ç–∞—Ä–∞–π—Å—è –æ—Å—Ç–∞—Ç—å—Å—è –ø–æ–ª–µ–∑–Ω—ã–º.
"""

print(rag_answer("–ö—Ç–æ –æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω—ã–π –∑–∞ –æ—Ä–≥–∞–Ω–∏–∑–∞—Ü–∏—é —Ñ—É—Ç–±–æ–ª–∞?", instruction))
```

```python
import os
import pandas as pd

TABLES_PATH = "data/–¢–∞–±–ª–∏—Ü—ã"

def load_table(table_name: str) -> pd.DataFrame:
    """–ó–∞–≥—Ä—É–∂–∞–µ—Ç —Ç–∞–±–ª–∏—Ü—É Excel –ø–æ –∏–º–µ–Ω–∏ —Ñ–∞–π–ª–∞ –±–µ–∑ —Ä–∞—Å—à–∏—Ä–µ–Ω–∏—è."""
    path = os.path.join(TABLES_PATH, f"{table_name}.xlsx")
    if not os.path.exists(path):
        raise FileNotFoundError(f"–§–∞–π–ª {path} –Ω–µ –Ω–∞–π–¥–µ–Ω.")

    df = pd.read_excel(path)
    df.columns = [c.strip() for c in df.columns]  # –æ—á–∏—â–∞–µ–º –ø—Ä–æ–±–µ–ª—ã –≤ –∑–∞–≥–æ–ª–æ–≤–∫–∞—Ö
    return df
```

```python
from pydantic import BaseModel, Field
from typing import Optional, Dict, Any
from typing import Union, List

class QueryCompanyData(BaseModel):
    """–ü–æ–∑–≤–æ–ª—è–µ—Ç –∏—Å–∫–∞—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –≤ –∫–æ—Ä–ø–æ—Ä–∞—Ç–∏–≤–Ω—ã—Ö —Ç–∞–±–ª–∏—Ü–∞—Ö –û–ë–ò–¢:
    –æ—Ñ–∏—Å—ã, –∫–æ–Ω—Ç–∞–∫—Ç—ã, —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∏, –∫–æ—Ä–ø–æ—Ä–∞—Ç–∏–≤—ã –∏ –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏.
    """

    table: str = Field(description="–ù–∞–∑–≤–∞–Ω–∏–µ —Ç–∞–±–ª–∏—Ü—ã: '–ê–¥—Ä–µ—Å–∞ –æ—Ñ–∏—Å–æ–≤', 'HR —Ä–∞–±–æ—Ç–∞ —Å –ø–µ—Ä—Å–æ–Ω–∞–ª–æ–º', '–ë–∞–∑–∞ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∏', '–ö–æ—Ä–ø–æ—Ä–∞—Ç–∏–≤—ã', '–ê–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏'")
    filters: Optional[Dict[str, Any]] = Field(default=None, description="–§–∏–ª—å—Ç—Ä—ã –ø–æ –∫–æ–ª–æ–Ω–∫–∞–º, –Ω–∞–ø—Ä–∏–º–µ—Ä {'–ì–æ—Ä–æ–¥':'–ú–æ—Å–∫–≤–∞'} –∏–ª–∏ {'–û—Ñ–∏—Å':'–õ–∏–≥–æ–≤—Å–∫–∏–π'}")
    what_to_return: Optional[Union[str, List[str]]] = Field(default=None, description="–ö–∞–∫–∏–µ —Å—Ç–æ–ª–±—Ü—ã –≤–µ—Ä–Ω—É—Ç—å: —Å—Ç—Ä–æ–∫–∞ –∏–ª–∏ —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫, –Ω–∞–ø—Ä–∏–º–µ—Ä '–¢–µ–ª–µ—Ñ–æ–Ω' –∏–ª–∏ ['–§–ò–û','–ö–æ–Ω—Ç–∞–∫—Ç']")

    def process(self, session_id: str = None):
        try:
            df = load_table(self.table)
        except Exception as e:
            return f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –∑–∞–≥—Ä—É–∑–∫–µ —Ç–∞–±–ª–∏—Ü—ã: {e}"

        # –ø—Ä–∏–º–µ–Ω—è–µ–º —Ñ–∏–ª—å—Ç—Ä—ã
        if self.filters:
            for k, v in self.filters.items():
                if k not in df.columns:
                    return f"–í —Ç–∞–±–ª–∏—Ü–µ '{self.table}' –Ω–µ—Ç —Å—Ç–æ–ª–±—Ü–∞ '{k}'. –î–æ—Å—Ç—É–ø–Ω—ã–µ: {', '.join(df.columns)}"
                df = df[df[k].astype(str).str.contains(str(v), case=False, na=False)]

        if df.empty:
            return f"–ü–æ –∑–∞–ø—Ä–æ—Å—É –Ω–∏—á–µ–≥–æ –Ω–µ –Ω–∞–π–¥–µ–Ω–æ –≤ —Ç–∞–±–ª–∏—Ü–µ '{self.table}'."

        # –≤—ã–±–∏—Ä–∞–µ–º –Ω—É–∂–Ω—ã–µ –ø–æ–ª—è
        if self.what_to_return:
            if isinstance(self.what_to_return, str):
                cols = [c for c in df.columns if self.what_to_return.lower() in c.lower()]
            elif isinstance(self.what_to_return, list):
                cols = [c for c in df.columns if any(w.lower() in c.lower() for w in self.what_to_return)]
            else:
                cols = []
            if cols:
                df = df[cols]
        return df.to_dict(orient="records")
```

```python
import json
import os
from pydantic import BaseModel, Field
from typing import Union

PROFILES_DIR = "user_profiles"
os.makedirs(PROFILES_DIR, exist_ok=True)


class CreateUserProfile(BaseModel):
    """–°–æ–∑–¥–∞–µ—Ç –ø—Ä–æ—Ñ–∏–ª—å –Ω–æ–≤–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è. –ú–æ–∂–Ω–æ –ø—Ä–æ–ø—É—Å—Ç–∏—Ç—å –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ, —É–∫–∞–∑–∞–≤ skip=True."""

    nickname: str | None = Field(default=None, description="–ò–º—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∏–ª–∏ –Ω–∏–∫")
    position: str | None = Field(default=None, description="–î–æ–ª–∂–Ω–æ—Å—Ç—å —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∞")
    location: str | None = Field(default=None, description="–ì–æ—Ä–æ–¥ –∏–ª–∏ –æ—Ñ–∏—Å")
    hobbies: list[str] | str | None = Field(default=None, description="–ò–Ω—Ç–µ—Ä–µ—Å—ã –∏–ª–∏ –≤–∏–¥—ã —Å–ø–æ—Ä—Ç–∞")
    status: str | None = Field(default=None, description="–°–æ—Ç—Ä—É–¥–Ω–∏–∫ –∏–ª–∏ –∫–∞–Ω–¥–∏–¥–∞—Ç")
    skip: bool = Field(default=False, description="–ü—Ä–æ–ø—É—Å—Ç–∏—Ç—å —Å–æ–∑–¥–∞–Ω–∏–µ –ø—Ä–æ—Ñ–∏–ª—è")

    def process(self, session_id: str):
        profile_path = os.path.join(PROFILES_DIR, f"{session_id}.json")

        # –µ—Å–ª–∏ –ø—Ä–æ—Ñ–∏–ª—å —É–∂–µ –µ—Å—Ç—å –∏ –ø–æ–º–µ—á–µ–Ω –∫–∞–∫ skipped ‚Äî –Ω–∏—á–µ–≥–æ –Ω–µ –¥–µ–ª–∞–µ–º
        if os.path.exists(profile_path):
            with open(profile_path, encoding="utf-8") as f:
                data = json.load(f)
            if data.get("skipped"):
                return "–¢—ã —É–∂–µ —Ä–µ—à–∏–ª –Ω–µ –∑–∞–ø–æ–ª–Ω—è—Ç—å –ø—Ä–æ—Ñ–∏–ª—å ‚Äî –≤—Å—ë –≤ –ø–æ—Ä—è–¥–∫–µ üôÇ"

        # –µ—Å–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —è–≤–Ω–æ –ø—Ä–æ–ø—É—Å–∫–∞–µ—Ç –∏–ª–∏ –Ω–∏—á–µ–≥–æ –Ω–µ —É–∫–∞–∑–∞–ª
        if self.skip or not any([self.nickname, self.position, self.location, self.hobbies, self.status]):
            profile = {
                "nickname": None,
                "position": None,
                "location": None,
                "hobbies": [],
                "status": None,
                "created": __import__('datetime').datetime.now().isoformat(),
                "updated": __import__('datetime').datetime.now().isoformat(),
                "skipped": True
            }
            with open(profile_path, "w", encoding="utf-8") as f:
                json.dump(profile, f, ensure_ascii=False, indent=2)
            return "–û–∫–µ–π –ø—Ä–æ—Ñ–∏–ª—å –º–æ–∂–Ω–æ –∑–∞–ø–æ–ª–Ω–∏—Ç—å –ø–æ–∑–∂–µ ‚Äî —è –ø–æ–º–æ–≥—É –∏ –±–µ–∑ –Ω–µ–≥–æ!"

        # –æ–±—ã—á–Ω–æ–µ —Å–æ–∑–¥–∞–Ω–∏–µ –ø—Ä–æ—Ñ–∏–ª—è
        profile = self.model_dump()
        if isinstance(profile.get("hobbies"), str):
            profile["hobbies"] = [profile["hobbies"]]
        profile["created"] = profile["updated"] = __import__('datetime').datetime.now().isoformat()

        with open(profile_path, "w", encoding="utf-8") as f:
            json.dump(profile, f, ensure_ascii=False, indent=2)

        name = profile["nickname"] or "–¥—Ä—É–≥"
        return f"–ü—Ä–æ—Ñ–∏–ª—å —Å–æ–∑–¥–∞–Ω! –ü—Ä–∏—è—Ç–Ω–æ –ø–æ–∑–Ω–∞–∫–æ–º–∏—Ç—å—Å—è, {name}"



class UpdateUserProfile(BaseModel):
    """–û–±–Ω–æ–≤–ª—è–µ—Ç –ø—Ä–æ—Ñ–∏–ª—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è ‚Äî –Ω–∞–ø—Ä–∏–º–µ—Ä, –ø—Ä–∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏–∏ –∏–Ω—Ç–µ—Ä–µ—Å–æ–≤ –∏–ª–∏ –ø–æ–¥–ø–∏—Å–æ–∫."""

    nickname: str = Field(description="–ò–º—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è")
    subscriptions: list[str] | None = Field(default=None, description="–¢–µ–º—ã, –Ω–∞ –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –ø–æ–¥–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è")
    hobbies: list[str] | None = Field(default=None, description="–û–±–Ω–æ–≤–ª—ë–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ –∏–Ω—Ç–µ—Ä–µ—Å–æ–≤")

    def process(self, session_id: str):
        profile_path = os.path.join(PROFILES_DIR, f"{session_id}.json")
        if not os.path.exists(profile_path):
            return "–ü—Ä–æ—Ñ–∏–ª—å –Ω–µ –Ω–∞–π–¥–µ–Ω. –°–æ–∑–¥–∞–π –Ω–æ–≤—ã–π —á–µ—Ä–µ–∑ –∑–Ω–∞–∫–æ–º—Å—Ç–≤–æ."
        profile = json.load(open(profile_path, "r", encoding="utf-8"))
        for key, value in self.model_dump(exclude_none=True).items():
            if isinstance(value, list):
                profile[key] = list(set(profile.get(key, []) + value))
            else:
                profile[key] = value
        profile["updated"] = __import__('datetime').datetime.now().isoformat()
        json.dump(profile, open(profile_path, "w", encoding="utf-8"), ensure_ascii=False, indent=2)
        return f"–ü—Ä–æ—Ñ–∏–ª—å {profile['nickname']} –æ–±–Ω–æ–≤–ª—ë–Ω."


class ReadUserProfile(BaseModel):
    """–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –¥–∞–Ω–Ω—ã–µ –ø—Ä–æ—Ñ–∏–ª—è –¥–ª—è –ø–æ–¥–º–µ—à–∏–≤–∞–Ω–∏—è –≤ –∫–æ–Ω—Ç–µ–∫—Å—Ç –∞–≥–µ–Ω—Ç–∞."""

    nickname: str = Field(description="–ò–º—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è")

    def process(self, session_id: str):
        path = os.path.join(PROFILES_DIR, f"{session_id}.json")
        if not os.path.exists(path):
            return "–ü—Ä–æ—Ñ–∏–ª—å –Ω–µ –Ω–∞–π–¥–µ–Ω."
        return json.load(open(path, encoding="utf-8"))
```

```python
import json
from openai import OpenAI
from pydantic import BaseModel

client = OpenAI(
    base_url="https://rest-assistant.api.cloud.yandex.net/v1",
    api_key=os.getenv("YANDEX_API_KEY"),
    project=os.getenv("FOLDER_ID")
)

model = f"gpt://{os.getenv('FOLDER_ID')}/gpt-oss-120b/latest"


class Agent:
    def __init__(self, instruction, tools=[], session_id="default", model=model, tool_choice="auto"):
        self.instruction = instruction
        self.model = model
        self.tool_choice = tool_choice
        self.tool_map = {x.__name__: x for x in tools if issubclass(x, BaseModel)}
        self.tools = [
            self._create_tool_annot(x) for x in tools
        ]
        self.user_sessions = {}

    def _create_tool_annot(self, x):
        if issubclass(x, BaseModel):
            return {
                "type": "function",
                "name": x.__name__,
                "description": x.__doc__,
                "parameters": x.model_json_schema(),
            }

    def __call__(self, message, session_id="default"):
        s = self.user_sessions.get(session_id, {"last_reply_id": None, "history": []})
        s["history"].append({"role": "user", "content": message})
        
        # –ü–æ–¥–º–µ—à–∏–≤–∞–µ–º RAG –∫–æ–Ω—Ç–µ–∫—Å—Ç
        context = fetch_context(message)
        if not context.strip():
            context = "–ö–æ–Ω—Ç–µ–∫—Å—Ç –ø–æ –∑–∞–ø—Ä–æ—Å—É –Ω–µ –Ω–∞–π–¥–µ–Ω."
            
        prompt = (
            f"{self.instruction}\n\n"
            f"–ö–æ–Ω—Ç–µ–∫—Å—Ç –∏–∑ –±–∞–∑—ã (–∏—Å–ø–æ–ª—å–∑—É–π —Ç–æ–ª—å–∫–æ —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ, –Ω–µ –≤—ã–¥—É–º—ã–≤–∞–π —Ñ–∞–∫—Ç—ã):\n"
            f"{context}\n\n"
            f"–ó–∞–ø—Ä–æ—Å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è: {message}\n"
            f"–ï—Å–ª–∏ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –Ω–µ—Ç ‚Äî —Å–∫–∞–∂–∏ –æ–± —ç—Ç–æ–º —á–µ—Å—Ç–Ω–æ."
        )

        # –û—Ç–ø—Ä–∞–≤–ª—è–µ–º –≤ –º–æ–¥–µ–ª—å
        res = client.responses.create(
            model=self.model,
            store=True,
            tools=self.tools,
            tool_choice=self.tool_choice,
            previous_response_id=s["last_reply_id"],
            input=prompt,
        )

        tool_calls = [item for item in res.output if item.type == "function_call"]
        if tool_calls:
            for call in tool_calls:
                fn = self.tool_map[call.name]
                obj = fn.model_validate(json.loads(call.arguments))
                result = obj.process(session_id)
                if not isinstance(result, str):
                    result = json.dumps(result, ensure_ascii=False)
                out = [{
                    "type": "function_call_output",
                    "call_id": call.call_id,
                    "output": result
                }]
                res = client.responses.create(
                    model=self.model,
                    input=out,
                    tools=self.tools,
                    previous_response_id=res.id,
                    store=True,
                )

        s["last_reply_id"] = res.id
        s["history"].append({"role": "assistant", "content": res.output_text})
        self.user_sessions[session_id] = s
        return res.output_text
```

```python
ALLOWED = {"b","strong","i","em","u","s","del","a","code","blockquote"}

def sanitize_html_for_telegram(text: str) -> str:
    if not text:
        return ""

    # <br> –ø–µ—Ä–µ–Ω–æ—Å —Å—Ç—Ä–æ–∫–∏
    text = re.sub(r"<\s*br\s*/?\s*>", "\n", text, flags=re.I)

    # —É–±–∏—Ä–∞–µ–º –±–ª–æ—á–Ω—ã–µ —Ç–µ–≥–∏
    text = re.sub(r"</?\s*(ul|ol|li|p|div|span|html|body|head|h[1-6]|table|tr|td|th|hr)\b[^>]*>", "", text, flags=re.I)
    text = re.sub(r"<!DOCTYPE[^>]*>", "", text, flags=re.I)

    # —É–¥–∞–ª—è–µ–º –Ω–µ–ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ–º—ã–µ —Ç–µ–≥–∏, –Ω–æ –æ—Å—Ç–∞–≤–ª—è–µ–º —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ
    def strip_unknown_tags(m):
        tag = m.group(1).lower()
        if tag in ALLOWED:
            return m.group(0)
        return ""

    text = re.sub(r"</?([a-zA-Z0-9]+)(\s[^>]*)?>", strip_unknown_tags, text)

    # —ç–∫—Ä–∞–Ω–∏—Ä—É–µ–º —Å–ª—É—á–∞–π–Ω—ã–µ –æ–¥–∏–Ω–æ–∫–∏–µ —É–≥–ª–æ–≤—ã–µ —Å–∫–æ–±–∫–∏
    text = text.replace("< ", "&lt; ").replace(" >", "&gt;").replace("<<", "&lt;&lt;").replace(">>", "&gt;&gt;")

    return text.strip()
```

```python
# —Ö—ç–Ω–¥–ª–µ—Ä—ã
@router.message(CommandStart())
async def start(msg: Message):
    content = '–ü—Ä–∏–≤–µ—Ç!üëã\n\n–£–∂–µ —Å–∫–æ—Ä–æ —Ç—ã —Å—Ç–∞–Ω–µ—à—å —á–∞—Å—Ç—å—é –Ω–∞—à–µ–π –±–æ–ª—å—à–æ–π –∫–æ–º–∞–Ω–¥—ã.\n\n' \
    '–ü–µ—Ä–≤—ã–π –¥–µ–Ω—å –Ω–∞ –Ω–æ–≤–æ–π —Ä–∞–±–æ—Ç–µ - —ç—Ç–æ –≤—Å–µ–≥–¥–∞ –≤–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ! ' \
    '–ù–æ –º—ã –ø–æ–º–æ–∂–µ–º —Ç–µ–±–µ —Å–¥–µ–ª–∞—Ç—å –µ–≥–æ –ª–µ–≥–∫–∏–º –∏ –ø–æ-–Ω–∞—Å—Ç–æ—è—â–µ–º—É –≤–¥–æ—Ö–Ω–æ–≤–ª—è—é—â–∏–ºü§©.\n\n–í–ø–µ—Ä–µ–¥üöÄ'
    keyboard = to_main_menu_kb()
    await msg.answer(content, reply_markup=keyboard)

@router.message(F.text == "–ì–ª–∞–≤–Ω–æ–µ –º–µ–Ω—é")
async def main_menu(msg: Message) -> None:
    content = '–¢—ã –Ω–∞—Ö–æ–¥–∏—à—å—Å—è –≤ –≥–ª–∞–≤–Ω–æ–º –º–µ–Ω—é. –í—ã–±–µ—Ä–∏ –∏–Ω—Ç–µ—Ä–µ—Å—É—é—â–∏–π –ø—É–Ω–∫—Ç:'
    keyboard = main_menu_kb(msg.from_user.id)
    await msg.answer(content, reply_markup=keyboard)   
    
@router.callback_query(F.data == 'subscribe_activities')
async def subscribe_activities(call: CallbackQuery):
    #—Å–ø–∏—Å–æ–∫ –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–µ–π –∏–∑ —Ç–∞–±–ª–∏—Ü—ã
    events = pd.read_excel(EVENTS_PATH, engine='openpyxl')
    buttons = []
    for _, row in events.iterrows():
        activity_name = str(row.get("–ê–∫—Ç–∏–≤–Ω–æ—Å—Ç—å ", "")).strip()
        buttons.append([InlineKeyboardButton(text=activity_name, callback_data=f"subscribe_{activity_name}")])
    buttons.append([InlineKeyboardButton(text="üîô –í –≥–ª–∞–≤–Ω–æ–µ –º–µ–Ω—é", callback_data="main_menu")])
    reply_markup = InlineKeyboardMarkup(inline_keyboard=buttons)
    await call.message.answer("–í—ã–±–µ—Ä–∏—Ç–µ –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –¥–ª—è –ø–æ–¥–ø–∏—Å–∫–∏:", reply_markup=reply_markup)

#–û–±—Ä–∞–±–æ—Ç–∫–∞ –ø–æ–¥–ø–∏—Å–∫–∏ –Ω–∞ –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å
@router.callback_query(F.data.startswith('subscribe_'))
async def handle_subscribe(call: CallbackQuery):
    activity = call.data[len('subscribe_'):]
    user_id = str(call.from_user.id)
    profile_path = os.path.join(PROFILES_DIR, f"{user_id}.json")
    if not os.path.isfile(profile_path):
        await call.answer("–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –∑–∞–ø–æ–ª–Ω–∏—Ç–µ –ø—Ä–æ—Ñ–∏–ª—å –ø–µ—Ä–µ–¥ –ø–æ–¥–ø–∏—Å–∫–æ–π.", show_alert=True)
        return

    with open(profile_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    subscriptions = data.get('subscriptions', [])
    if activity not in subscriptions:
        subscriptions.append(activity)
        data['subscriptions'] = subscriptions
        with open(profile_path, "w", encoding="utf-8") as f:
            json.dump(data, f, ensure_ascii=False, indent=2)
        await call.answer(f"–ü–æ–¥–ø–∏—Å–∫–∞ –Ω–∞ '{activity}' –¥–æ–±–∞–≤–ª–µ–Ω–∞!", show_alert=True)
    else:
        await call.answer(f"–í—ã —É–∂–µ –ø–æ–¥–ø–∏—Å–∞–Ω—ã –Ω–∞ '{activity}'.", show_alert=True)

@router.callback_query(F.data == 'buddy')
async def answer_buddy(call: CallbackQuery):
    session_id = str(call.from_user.id)
    profile_path = os.path.join(PROFILES_DIR, f"{session_id}.json")

    # –µ—Å–ª–∏ –ø—Ä–æ—Ñ–∏–ª—å —É–∂–µ –µ—Å—Ç—å ‚Äî –ø—Ä–æ—Å—Ç–æ –≤—ã—Ö–æ–¥–∏–º –≤ —Ä–µ–∂–∏–º –æ–±—â–µ–Ω–∏—è
    if os.path.exists(profile_path):
        with open(profile_path, encoding="utf-8") as f:
            data = json.load(f)
        if data.get("skipped") or any([data.get("nickname"), data.get("position"), data.get("location"),
                                       data.get("hobbies"), data.get("status")]):
            await bot.send_message(
                call.from_user.id,
                "–°–Ω–æ–≤–∞ –ø—Ä–∏–≤–µ—Ç \n–ú–æ–∂–µ—à—å –∑–∞–¥–∞—Ç—å –º–Ω–µ –≤–æ–ø—Ä–æ—Å"
            )
            return

    # –µ—Å–ª–∏ –ø—Ä–æ—Ñ–∏–ª—è –Ω–µ—Ç ‚Äî –ø—Ä–µ–¥–ª–∞–≥–∞–µ–º –ø–æ–∑–Ω–∞–∫–æ–º–∏—Ç—å—Å—è
    greeting = (
        "–ü—Ä–∏–≤–µ—Ç! –Ø —Ç–≤–æ–π –∫–æ—Ä–ø–æ—Ä–∞—Ç–∏–≤–Ω—ã–π –ë–∞–¥–¥–∏ \n\n"
        "–î–∞–≤–∞–π –ø–æ–∑–Ω–∞–∫–æ–º–∏–º—Å—è, —á—Ç–æ–±—ã —è –º–æ–≥ –ø–æ–º–æ–≥–∞—Ç—å —Ç–µ–±–µ —Ç–æ—á–Ω–µ–µ.\n\n"
        "–†–∞—Å—Å–∫–∞–∂–∏ –æ —Å–µ–±–µ (–µ—Å–ª–∏ —Ö–æ—á–µ—à—å):\n"
        "1Ô∏è‚É£ –ö–∞–∫ —Ç–µ–±—è –∑–æ–≤—É—Ç\n"
        "2Ô∏è‚É£ –¢–≤–æ—è –¥–æ–ª–∂–Ω–æ—Å—Ç—å (–∏–ª–∏ –æ—Ç–¥–µ–ª)\n"
        "3Ô∏è‚É£ –í –∫–∞–∫–æ–º –≥–æ—Ä–æ–¥–µ –∏–ª–∏ –æ—Ñ–∏—Å–µ —Ä–∞–±–æ—Ç–∞–µ—à—å\n"
        "4Ô∏è‚É£ –õ—é–±–∏–º—ã–π —Å–ø–æ—Ä—Ç –∏–ª–∏ —Ö–æ–±–±–∏\n"
        "5Ô∏è‚É£ –¢—ã —É–∂–µ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫ –∏–ª–∏ –∫–∞–Ω–¥–∏–¥–∞—Ç\n\n"
        "–ï—Å–ª–∏ –Ω–µ —Ö–æ—á–µ—à—å —Å–µ–π—á–∞—Å ‚Äî –º–æ–∂–µ—à—å –ø—Ä–æ–ø—É—Å—Ç–∏—Ç—å."
    )

    kb = InlineKeyboardMarkup(
        inline_keyboard=[
            [InlineKeyboardButton(text="–ü—Ä–æ–ø—É—Å—Ç–∏—Ç—å", callback_data="skip_profile")]
        ]
    )
    await bot.send_message(call.from_user.id, greeting, reply_markup=kb)

@router.callback_query(F.data == "skip_profile")
async def skip_profile(call: CallbackQuery):
    session_id = str(call.from_user.id)
    profile_path = os.path.join(PROFILES_DIR, f"{session_id}.json")

    skipped_profile = {
        "nickname": None,
        "position": None,
        "location": None,
        "hobbies": [],
        "status": None,
        "created": datetime.now().isoformat(),
        "updated": datetime.now().isoformat(),
        "skipped": True
    }

    with open(profile_path, "w", encoding="utf-8") as f:
        json.dump(skipped_profile, f, ensure_ascii=False, indent=2)

    await call.message.edit_text("–û–∫–µ–π —è –≤—Å—ë —Ä–∞–≤–Ω–æ –ø–æ–º–æ–≥—É —á–µ–º —Å–º–æ–≥—É. –ó–∞–¥–∞–≤–∞–π –≤–æ–ø—Ä–æ—Å!")    
    
@router.message()
async def handle_user_message(msg: Message):
    session_id = str(msg.from_user.id)
    query = msg.text
    await bot.send_chat_action(msg.chat.id, action="typing")  # —ç—Ñ—Ñ–µ–∫—Ç ¬´–ø–µ—á–∞—Ç–∞–µ—Ç‚Ä¶¬ª

    try:
        # –≤—ã–∑—ã–≤–∞–µ–º –∞–≥–µ–Ω—Ç–∞
        answer = buddy_agent(query, session_id=session_id)
        clean_answer = sanitize_html_for_telegram(answer)
        await msg.answer(clean_answer)
    except Exception as e:
        await msg.answer(f"–ß—Ç–æ-—Ç–æ –ø–æ—à–ª–æ –Ω–µ —Ç–∞–∫:\n{e}")

@router.callback_query(F.data == 'buddy_can')
async def answer_buddy_can(call: CallbackQuery):
    content = '–Ø –º–æ–≥—É –ø–æ–º–æ—á—å —Ç–µ–±–µ:\n' \
    '‚Ä¢ <b>–ó–∞–∫–∞–∑–∞—Ç—å —Å–ø—Ä–∞–≤–∫—É 2-–ù–î–§–õ</b> (–≤ —Ä–∞–∑—Ä–∞–±–æ—Ç–∫–µ)\n' \
    '‚Ä¢ <b>–°–∫–∞–∑–∞—Ç—å ¬´–°–ø–∞—Å–∏–±–æ¬ª –∫–æ–ª–ª–µ–≥–µ</b> (–≤ —Ä–∞–∑—Ä–∞–±–æ—Ç–∫–µ)\n' \
    '‚Ä¢ <b>–ü–æ–¥–ø–∏—Å–∞—Ç—å –Ω–∞ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–µ</b> ‚Äî –∑–∞—Ä–µ–≥–∏—Å—Ç—Ä–∏—Ä—É—é –∏ –ø—Ä–∏—à–ª—é –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ\n' \
    '‚Ä¢ <b>–î–∞—Ç—å –∫–æ–Ω—Ç–∞–∫—Ç—ã –Ω—É–∂–Ω–æ–≥–æ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∞</b> –∏–ª–∏ –ø–æ–¥—Å–∫–∞–∑–∞—Ç—å –∞–¥—Ä–µ—Å –æ—Ñ–∏—Å–∞\n' \
    '‚Ä¢ <b>–†–∞—Å—Å–∫–∞–∑–∞—Ç—å –≤—Å—ë, —á—Ç–æ —è –∑–Ω–∞—é –æ –Ω–∞—à–µ–π –∫–æ–º–ø–∞–Ω–∏–∏</b> ‚Äî –æ—Ç –º–∏—Å—Å–∏–∏ –¥–æ –ø—Ä–∞–≤–∏–ª –∏ –ø–æ–ª–µ–∑–Ω—ã—Ö —Å—Å—ã–ª–æ–∫'
    
    keyboard = to_main_menu_kb()
    await call.message.answer(content, reply_markup=keyboard)
        
@router.callback_query(
    lambda F: F.data.not_in({'buddy', 'fill_profile', 'buddy_can'}) & ~F.data.startswith('subscribe_')
)
async def answer_callback(call: CallbackQuery):
    content = '–¢—ã –Ω–∞—Ö–æ–¥–∏—à—å—Å—è –≤ –≥–ª–∞–≤–Ω–æ–º –º–µ–Ω—é. –í—ã–±–µ—Ä–∏ –∏–Ω—Ç–µ—Ä–µ—Å—É—é—â–∏–π –ø—É–Ω–∫—Ç:'
    keyboard = main_menu_kb(call.from_user.id)
    await call.message.answer(content, reply_markup=keyboard)
    
# –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ—Ñ–∏–ª—è
@router.callback_query(F.data == 'fill_profile')
async def fill_nickname(call: CallbackQuery, state: FSMContext):
    content = '–ö–∞–∫ —Ç–µ–±—è –∑–æ–≤—É—Ç?'
    await state.set_state(FillProfile.position)
    keyboard = to_main_menu_kb()
    await call.message.answer(content, reply_markup=keyboard)

@router.message(FillProfile.position)
async def fill_position(msg: Message, state: FSMContext):
    content = '–ö–∞–∫–∞—è —É —Ç–µ–±—è –¥–æ–ª–∂–Ω–æ—Å—Ç—å –∏–ª–∏ –æ—Ç–¥–µ–ª?'
    await state.update_data(nickname=msg.text)
    await state.set_state(FillProfile.location)
    keyboard = to_main_menu_kb()
    await msg.answer(content, reply_markup=keyboard)

@router.message(FillProfile.location)
async def fill_location(msg: Message, state: FSMContext):
    content = '–í –∫–∞–∫–æ–º –≥–æ—Ä–æ–¥–µ –∏–ª–∏ –æ—Ñ–∏—Å–µ —Ç—ã —Ä–∞–±–æ—Ç–∞–µ—à—å?'
    await state.update_data(position=msg.text)
    await state.set_state(FillProfile.hobbies)
    keyboard = to_main_menu_kb()
    await msg.answer(content, reply_markup=keyboard)

@router.message(FillProfile.hobbies)
async def fill_hobbies(msg: Message, state: FSMContext):
    content = '–õ—é–±–∏–º—ã–π —Å–ø–æ—Ä—Ç –∏–ª–∏ —Ö–æ–±–±–∏?'
    await state.update_data(location=msg.text)
    await state.set_state(FillProfile.status)
    keyboard = to_main_menu_kb()
    await msg.answer(content, reply_markup=keyboard)

@router.message(FillProfile.status)
async def fill_status(msg: Message, state: FSMContext):
    content = '–¢—ã —É–∂–µ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫ –∏–ª–∏ –∫–∞–Ω–¥–∏–¥–∞—Ç?'
    await state.update_data(hobbies=msg.text)
    await state.set_state(FillProfile.end_fill)
    keyboard = to_main_menu_kb()
    await msg.answer(content, reply_markup=keyboard)

@router.message(FillProfile.end_fill)
async def end_fill(msg: Message, state: FSMContext):
    user_id = str(msg.from_user.id)
    profile_path = os.path.join(PROFILES_DIR, f"{user_id}.json")

    data = await state.get_data()
    data['status'] = msg.text
    data['skip'] = False
    data['created'] = datetime.now().isoformat()
    data['updated'] = datetime.now().isoformat()

    with open(profile_path, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=2)

    if await msg.answer("–û—Ç–ª–∏—á–Ω–æ! üëè –ü—Ä–æ—Ñ–∏–ª—å —Å–æ—Ö—Ä–∞–Ω—ë–Ω. –¢–µ–ø–µ—Ä—å —è –º–æ–≥—É –ø–æ–º–æ–≥–∞—Ç—å —Ç–µ–±–µ —Ç–æ—á–Ω–µ–µ."):
        await state.clear()
        await main_menu(msg)
```
